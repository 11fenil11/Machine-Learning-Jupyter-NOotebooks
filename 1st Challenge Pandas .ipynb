```python
import pandas as pd
```

# For CLIENTDATA.CSV

```python
cli = pd.read_csv('E:\Course\clientdata.csv')
cli.head()
```

```python
cli.dtypes
```

```python
cli['BillDate'] = pd.to_datetime(cli.BillDate)
```

```python
cli.head()
```

```python
cli['day'] = cli.BillDate.dt.weekday_name
```

```python
cli.head()
```

```python
cli.groupby('day').sum()
```

# Total Sale categorized by day wise

```python
cli.groupby('day').BillRate.sum()
```

# Average sale per day

```python
cli.groupby('day').BillRate.mean()
```
